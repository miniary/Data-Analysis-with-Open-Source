```python
import requests

url = 'http://openapi.seoul.go.kr:8088/6a704e736f6f6f6f393352466c5448/json/energyUseDataSummaryInfo/1/5/2015/01'
api_key = '6a704e736f6f6f6f393352466c5448'

# 연도, 월
years = range(2015, 2025)
months = range(1, 13)

# 연도와 월 입력 반복문
for year in years :
    for month in months :
        url = f'http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/5/{year}/{month}'

        params = {
            'KEY': '6a704e736f6f6f6f393352466c5448',
            'TYPE': 'json',
            'SERVICE': 'energyUseDataSummaryInfo',
            'START_INDEX': '1',
            'END_INDEX': '2'
        }


    ## 데이터 수집
        # print(url)
        response = requests.get(url, params=params)
        data = response.json()

    # ## 호출 성공/실패 출력
        if response.status_code == 200:
            print(data)
    #     print(data['energyUseDataSummaryInfo']['row'])
    #     # print(data['row'])
    #     #   print("서울시 종로구의 현재 온도는 : {0}{1} 입니다.".format(data['current']['temperature_2m'], data['current_units']['temperature_2m']))
    # else:
    #     print(f"Error: API returned status code {response.status_code}")
    #     print(response.text)
```
