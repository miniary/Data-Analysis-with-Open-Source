```python
import requests

api_key = '6a704e736f6f6f6f393352466c5448'

# 연도, 월
years = range(2015, 2025)
months = range(1, 13)

# 연도와 월 입력 반복문 사용 URL 생성
for year in years :
    for month in months :
        url = f'http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/100/{year}/{month:02d}'

        params = {
            'KEY': '6a704e736f6f6f6f393352466c5448',
            'TYPE': 'json',
            'SERVICE': 'energyUseDataSummaryInfo'
        }

    ## 데이터 수집
        response = requests.get(url, params=params)
        data = response.json()

    # ## 호출 성공/실패 출력
        if response.status_code == 200:
            rows = data['energyUseDataSummaryInfo']['row']

            for r in rows :
                if r['MM_TYPE'] == '개인':
                    print({
                        '년도' : r['YEAR'] ,
                        '월' : r['MON'] ,
                        '회원유형' : r['MM_TYPE'],
                        '현년도 전기 사용량' :  r['EUS'],
                        '현년도 가스 사용량' :  r['GUS'],
                        '현년도 수도 사용량' :  r['WUS'],
                        '현년도 지역난방 에너지 사용량' :   r['HUS']
                    })
        else:
            print(f"Error: API 호출 실패 : {response.status_code}")
```
